```python
file = 'raw_analyst_ratings.csv'
print(f"Analyzing {file}")
# Create the full file path
file_path = os.path.join('data', 'raw_analyst', file)

# Check if file exists
if not os.path.exists(file_path):
    print(f"File not found: {file_path}")
```

```python
# Create an instance of the EDA class
analysis = NewsAnalysis(file_path)
```

## Descriptive Statistics

# Headline Length Calculation
Computes the length of each headline and adds it as a new column, useful for understanding the distribution of headline lengths.

```python
analysis.calculate_headline_length()
```

# Articles per Publisher
Counts the number of articles contributed by each publisher to identify the most active publishers.

```python
publisher_counts = analysis.articles_per_publisher()
print(publisher_counts)
```

# Publication Trends

## Text Analysis (Sentiment Analysis & Topic Modeling):

Analyzes the sentiment of each headline to classify them into positive, neutral, or negative sentiments. It also plots the sentiment distribution.

```python
sentiments = analysis.sentiment_analysis()
print(sentiments)
```
