```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
print(os.getcwd())  # This prints the current working directory
os.chdir('c:\\Financial-News-Sentiment-Analysis')  # Set the working directory to the project root

# Import the EDA class
from scripts.Stock_Data_EDA import EDA

# Import the display function
from IPython.display import display, HTML
```

## List Of Data 

```python

files = [
    'AAPL_historical_data.csv',
    'AMZN_historical_data.csv',
    'GOOG_historical_data.csv',
    'META_historical_data.csv',
    'MSFT_historical_data.csv',
    'NVDA_historical_data.csv',
    'TSLA_historical_data.csv'
]
```

# Loop through each File and Perform EDA

```python

for file_name in files:
    print(f"Analyzing {file_name}")
    display(HTML(f"<h1>Analyzing {file_name}</h1>"))
    
    # Create the full file path
    file_path = os.path.join('data', 'yfinance_data', file_name)
    
    # Check if file exists
    if not os.path.exists(file_path):
        print(f"File not found: {file_path}")
        continue
    
    # Create an instance of the EDA class
    eda = EDA(file_path)
    
    # Load the data
    eda.load_data()
    
    # Perform EDA operations
    
    display(HTML(f"<h1>Descriptive Statistics for {file_name}</h1>"))
    display(HTML(f"<h3>This section provides summary statistics such as mean, standard deviation, min, and max for the dataset.</h3>"))
    eda.descriptive_statistics()

    display(HTML(f"<h2>Number of Data Points per Year for {file_name}</h2>"))
    display(HTML(f"<h3>This section counts the number of entries per year, helping to identify the dataset's coverage over time.</h3>"))
    eda.count_data_points_per_year()

    display(HTML(f"<h2>Date Trends for {file_name}</h2>"))
    display(HTML(f"<h3>This section visualizes the trend of data entries over the years.</h3>"))
    eda.plot_date_trends()

    display(HTML(f"<h2>Stock Prices for {file_name}</h2>"))
    display(HTML(f"<h3>This section plots the closing stock prices over time, showing how the stock's value has changed.</h3>"))
    eda.plot_stock_prices()

    display(HTML(f"<h2>Volume for {file_name}</h2>"))
    display(HTML(f"<h3>This section visualizes the trading volume over time, indicating the activity level of the stock.</h3>"))
    eda.plot_volume()

    display(HTML(f"<h2>Correlation Heatmap for {file_name}</h2>"))
    display(HTML(f"<h3>This heatmap shows the correlation between different variables, revealing relationships in the dataset.</h3>"))
    eda.plot_correlation_heatmap()

    display(HTML(f"<h2>Relative Strength Index (RSI) for {file_name}</h2>"))
    display(HTML(f"<h3>This section calculates and visualizes the RSI, an indicator used to assess stock price momentum.</h3>"))
    eda.plot_calculate_rsi()

    display(HTML(f"<h2>Moving Average Convergence Divergence (MACD) for {file_name}</h2>"))
    display(HTML(f"<h3>This section calculates and visualizes the MACD, an indicator that shows the relationship between two moving averages of a stock's price.</h3>"))
    eda.plot_calculate_macd()

```
