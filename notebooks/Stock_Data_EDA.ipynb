```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
print(os.getcwd())  # This prints the current working directory
os.chdir('c:\\Financial-News-Sentiment-Analysis')  # Set the working directory to the project root

# Import the EDA class
from scripts.Descriptive_Statistics import EDA
```

## List Of Data 

```python
files = [
    'AAPL_historical_data.csv',
    'AMZN_historical_data.csv',
    'GOOG_historical_data.csv',
    'META_historical_data.csv',
    'MSFT_historical_data.csv',
    'NVDA_historical_data.csv',
    'TSLA_historical_data.csv'
]
```

# Loop through each file and perform EDA

```python

for file_name in files:
    print(f"Analyzing {file_name}")
    
    # Create the full file path
    file_path = os.path.join('data', 'yfinance_data', file_name)
    
    # Check if file exists
    if not os.path.exists(file_path):
        print(f"File not found: {file_path}")
        continue
    
    # Create an instance of the EDA class
    eda = EDA(file_path)
    
    # Load the data
    eda.load_data()
    
    # Perform EDA operations
    print(f"Descriptive Statistics for {file_name}")
    eda.descriptive_statistics()
    print(f"Number of Data Points per Year for {file_name}")
    eda.count_data_points_per_year()
    print(f"Date Trends for {file_name}")
    eda.plot_date_trends()
    print(f"Stock Prices for {file_name}")
    eda.plot_stock_prices()
    print(f"Volume for {file_name}")
    # eda.plot_volume()
    plt.figure(figsize=(10, 6))
    plt.plot(eda.df["Date"], eda.df["Volume"])
    plt.title("Trading Volume Over Time")
    plt.xlabel("Date")
    plt.ylabel("Volume")
    plt.ticklabel_format(style="plain", axis="y")
    plt.legend()
    plt.show()
    print(f"Correlation Heatmap for {file_name}")
    # eda.plot_correlation_heatmap()
    plt.figure(figsize=(10, 8))
    sns.heatmap(eda.df.corr(), annot=True, cmap="coolwarm")
    plt.title("Correlation Heatmap")
    plt.show()
```
